```python
# Install necessary libraries
!pip install ics pandas

# Import libraries
import requests
from ics import Calendar
from datetime import datetime
import pandas as pd

# URLs to .ics event calendars
urls = {
    "Python Events": "https://www.google.com/calendar/ical/j7gov1cmnqr9tvg14k621j7t5c%40group.calendar.google.com/public/basic.ics",
    "User Group Events": "https://www.google.com/calendar/ical/3haig2m9msslkpf2tn1h56nn9g%40group.calendar.google.com/public/basic.ics"
}

# Current timestamp to record fetch time
fetch_time = datetime.utcnow().strftime("%Y-%m-%d %H:%M:%S UTC")

# List to hold DataFrames for each calendar
dfs = []

# Process each calendar
for label, url in urls.items():
    print(f"\n📥 Fetching events from: {label}")
    response = requests.get(url)

    if response.status_code == 200:
        calendar = Calendar(response.text)

        event_data = []

        for event in calendar.events:
            event_data.append({
                "calendar": label,
                "name": event.name,
                "start": event.begin.format("YYYY-MM-DD HH:mm") if event.begin else None,
                "end": event.end.format("YYYY-MM-DD HH:mm") if event.end else None,
                "location": event.location if event.location else "N/A",
                "description": event.description[:200] + "..." if event.description else "N/A",
                "fetched_at": fetch_time
            })

        df = pd.DataFrame(event_data)
        dfs.append(df)

        # Display first few rows as a sample
        print(df.head())
    else:
        print(f"❌ Failed to fetch {label} — Status code: {response.status_code}")

# Combine all dataframes into one
all_events_df = pd.concat(dfs, ignore_index=True)

# Save to CSV
csv_filename = f"python_events_{datetime.utcnow().strftime('%Y%m%d_%H%M%S')}.csv"
all_events_df.to_csv(csv_filename, index=False)
print(f"\n✅ CSV saved as: {csv_filename}")

# Optional: Display summary
print(f"\n📊 Total events fetched: {len(all_events_df)}")
print(f"🕒 Data last fetched at: {fetch_time}")
```
