```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
import time
import pandas as pd

# Set your ChromeDriver path
chrome_path = r"C:\Users\CARBON\Downloads\chromedriver-win64\chromedriver.exe"
service = Service(chrome_path)
driver = webdriver.Chrome(service=service)

# Open Eventbrite page for Toronto weekend events
url = "https://www.eventbrite.ca/d/canada--toronto/events--this-weekend/"
driver.get(url)

# Wait for content to load
time.sleep(5)

# Scroll to load more (optional)
driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
time.sleep(3)

# Extract event containers
cards = driver.find_elements(By.CLASS_NAME, "eds-event-card-content__content")

# Collect data
events = []

for card in cards:
    try:
        title = card.find_element(By.CLASS_NAME, "eds-event-card__formatted-name--is-clamped").text
    except:
        title = "N/A"

    try:
        date = card.find_element(By.CLASS_NAME, "eds-text-bs--fixed").text
    except:
        date = "N/A"

    try:
        location = card.find_element(By.CLASS_NAME, "card-text--truncated__one").text
    except:
        location = "N/A"

    try:
        link = card.find_element(By.TAG_NAME, "a").get_attribute("href")
    except:
        link = "N/A"

    events.append({
        "Event Name": title,
        "Date": date,
        "Location": location,
        "URL": link
    })

# Close browser
driver.quit()

# Save to CSV
df = pd.DataFrame(events)
df.to_csv("eventbrite_scraped_events.csv", index=False)
print(f"✅ Done! Saved {len(df)} events to 'eventbrite_scraped_events.csv'")
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
import pandas as pd

# ChromeDriver path
chrome_path = r"C:\Users\CARBON\Downloads\chromedriver-win64\chromedriver.exe"
service = Service(chrome_path)
driver = webdriver.Chrome(service=service)

# Load Eventbrite page
url = "https://www.eventbrite.ca/d/canada--toronto/events--this-weekend/"
driver.get(url)

# ✅ Wait up to 15 seconds for event cards to appear
try:
    WebDriverWait(driver, 15).until(
        EC.presence_of_element_located((By.CLASS_NAME, "eds-event-card-content__content"))
    )
    print("✅ Events loaded")
except:
    print("❌ Event cards did not load in time")

# Scroll to load more (optional)
driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
time.sleep(3)

# Extract event cards
cards = driver.find_elements(By.CLASS_NAME, "eds-event-card-content__content")
print(f"🎯 Found {len(cards)} event cards")

events = []

for card in cards:
    try:
        title = card.find_element(By.CLASS_NAME, "eds-event-card__formatted-name--is-clamped").text
    except:
        title = "N/A"

    try:
        date = card.find_element(By.CLASS_NAME, "eds-text-bs--fixed").text
    except:
        date = "N/A"

    try:
        location = card.find_element(By.CLASS_NAME, "card-text--truncated__one").text
    except:
        location = "N/A"

    try:
        link = card.find_element(By.TAG_NAME, "a").get_attribute("href")
    except:
        link = "N/A"

    events.append({
        "Event Name": title,
        "Date": date,
        "Location": location,
        "URL": link
    })

driver.quit()

# Save to CSV
df = pd.DataFrame(events)
df.to_csv("eventbrite_scraped_events.csv", index=False)
print(f"✅ Done! Saved {len(df)} events to CSV.")
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
import pandas as pd

# ChromeDriver path
chrome_path = r"C:\Users\CARBON\Downloads\chromedriver-win64\chromedriver.exe"
service = Service(chrome_path)
driver = webdriver.Chrome(service=service)

# Load Eventbrite Toronto events
url = "https://www.eventbrite.ca/d/canada--toronto/events--this-weekend/"
driver.get(url)

# ✅ Scroll to trigger loading
driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
time.sleep(5)  # give time to load after scroll

# ✅ Wait until event wrapper loads (new class name)
try:
    WebDriverWait(driver, 15).until(
        EC.presence_of_element_located((By.CLASS_NAME, "search-event-card-wrapper"))
    )
    print("✅ Events loaded")
except:
    print("❌ Event cards still didn’t load")

# Get the event blocks
cards = driver.find_elements(By.CLASS_NAME, "search-event-card-wrapper")
print(f"🎯 Found {len(cards)} event cards")

events = []

for card in cards:
    try:
        title = card.find_element(By.CLASS_NAME, "event-card__formatted-name--content").text
    except:
        title = "N/A"

    try:
        date = card.find_element(By.CLASS_NAME, "event-card__sub-title").text
    except:
        date = "N/A"

    try:
        location = card.find_element(By.CLASS_NAME, "event-card__sub-content").text
    except:
        location = "N/A"

    try:
        link = card.find_element(By.TAG_NAME, "a").get_attribute("href")
    except:
        link = "N/A"

    events.append({
        "Event Name": title,
        "Date": date,
        "Location": location,
        "URL": link
    })

# Close browser
driver.quit()

# Save to CSV
df = pd.DataFrame(events)
df.to_csv("eventbrite_scraped_events.csv", index=False)
print(f"✅ Done! Saved {len(df)} events to CSV.")
```
